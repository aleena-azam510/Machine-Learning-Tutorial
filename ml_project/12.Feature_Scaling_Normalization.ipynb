### Normalization (Min-Max Scaling) in Feature Scaling

### Normalization, also known as Min-Max scaling, is a technique used in feature scaling to transform the range of features to a specific scale, typically between 0 and 1. This is achieved by linearly transforming the original data.   

### Formula:

### X_scaled = (X - X_min) / (X_max - X_min)
### where:

### X_scaled: The scaled value of the feature
### X: The original value of the feature
### X_min: The minimum value of the feature in the dataset
### X_max: The maximum value of the feature in the dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
mms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"]=mms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(dataset["CoapplicantIncome"])
plt.subplot(1,2,2)
sns.distplot(dataset["CoapplicantIncome_min"])
plt.title('After')



plt.show()
```
