```python
#  function transformer is a tool that applies a specific function to every element in a collection (like a list or array).

# Think of it like this:

# Imagine you have a basket of apples. You want to know the weight of each apple,so you use a scale (the function). A function transformer
# would be like having a magical machine that automatically weighs every apple in the basket for you, giving you a list of individual weights.
```

### Using log

```python
ft = FunctionTransformer(func=np.log1p)
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["tf_CoapplicantIncome"]=ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(dataset["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title('After')
sns.distplot(dataset["tf_CoapplicantIncome"])
plt.show()
```

### Using lambda

```python
dataset = pd.read_csv("loan.csv" )
dataset.head(3)
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
IQR = q3-q1
min_range = q1-(1.5*IQR)
max_range = q3+(1.5*IQR)
dataset=dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=lambda x:x**2)
ft.fit(dataset[["CoapplicantIncome"]])
dataset["tf2_CoapplicantIncome"]=ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(dataset["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title('After')
sns.distplot(dataset["tf2_CoapplicantIncome"])
plt.show()
```
