### Features are the individual characteristics or properties of a dataset that describe the objects or instances being studied. In other words, they are the columns in your data table. For example, in a dataset of cars, features might include:

# Make,Model,Year,Mileage,Price
### Feature selection is the process of choosing the most relevant features from a dataset to use in a machine learning model. 

### Forward Selection and Backward Elimination are two popular methods for feature selection in machine learning. They help us choose the most important features from a dataset to build better models.

### Forward Selection:

### Starts with an empty set of features.
### Adds one feature at a time.
### Selects the feature that improves the model's performance the most.
### Repeats until adding more features doesn't significantly improve performance.

### Backward Elimination:

### Starts with all features included.
### Removes one feature at a time.
### Removes the feature that has the least impact on the model's performance.
### Repeats until removing features significantly decreases performance.

```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset = pd.read_csv("diabetes.csv")
```

```python
dataset.head(3)
```

```python
X=dataset.iloc[:,:-1]
y=dataset["Outcome"]
```

```python
X.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
fs=SequentialFeatureSelector(lr,k_features=5,forward=True)  #for forward elimination it is true,for backward it is false
```

```python
fs.fit(X,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
