```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("placement.csv")
dataset.head(3)
```

```python
sns.scatterplot(x="cgpa",y="placement_exam_marks",data=dataset,hue="placed")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placed"]
```

# Logistic Regression (Binary Classification) (Polynomial input)
 # if data is not separable clearly



```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=3)
pf.fit(x)
x=pd.DataFrame(pf.transform(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict([[7.19, 26.0]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```
