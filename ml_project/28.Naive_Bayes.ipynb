```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv("placement.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x="cgpa",y="placement_exam_marks",hue="placed",data=df)
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test) , gnb.score(x_train,y_train)
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test) , mnb.score(x_train,y_train)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```
