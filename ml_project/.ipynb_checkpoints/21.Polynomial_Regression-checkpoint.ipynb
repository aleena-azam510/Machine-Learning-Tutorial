```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("ice_cream selling data.csv")
dataset.head(3)
```

```python
plt.scatter(dataset["Temperature (°C)"],dataset["Ice Cream Sales (units)"])
plt.show()
```

```python
dataset.corr()
```

```python
x = dataset[["Temperature (°C)"]]
y= dataset["Ice Cream Sales (units)"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=2)
```

```python
pf.fit(x)
```

```python
x=pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred=lr.predict(x)
```

```python
plt.scatter(dataset["Temperature (°C)"],dataset["Ice Cream Sales (units)"])
plt.plot(dataset["Temperature (°C)"],y_pred,c="red")
plt.show()
```
