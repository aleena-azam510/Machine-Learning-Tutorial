```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("rank_salary.csv")
df.head(3)
```

```python
x=df[["Level"]]
y=df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
```

```python
dc=DecisionTreeRegressor(criterion= 'squared_error',max_depth= 1,splitter='best')
dc.fit(x_train,y_train)
```

```python
dc.score(x_test,y_test)*100
```

```python
dc.score(x_train,y_train)*100
```

```python
param = {"criterion":["squared_error", "friedman_mse", "absolute_error","poisson"],
    "splitter":['best',"random"],
    "max_depth":[i for i in range(1,20)]}
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
gd=GridSearchCV(DecisionTreeRegressor(),param_grid=param)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_
```

```python
rd=RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=param,n_iter=20)
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```
