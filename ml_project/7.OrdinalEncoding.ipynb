```python
import pandas as pd
```

```python
df = pd.DataFrame({"size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
```

```python
df.head()
```

```python
ord_data = [["s","m","l","xl"]]
```

### Ordinal Encoding Using Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
```

```python
oe.fit(df[["size"]])
```

```python
df["size_scikit_enc"]=oe.transform(df[["size"]])
df
```

```python
ord_data1 = {"s":0 , "m":1 ,"l":2 ,"xl":3}
```

```python
df["size_map_enc"] = df["size"].map(ord_data1)
df
```
